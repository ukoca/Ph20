```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def compute_functions (f_x, f_y, A_x, A_y, phaseshift, time_interval, num_iterations, useNumpy):
    if (useNumpy):
        time_array = np.array([i for i in range(num_iterations)])
        x_t = A_x*np.cos(2*math.pi*f_x*time_array*time_interval)
        y_t = A_y*np.sin(2*math.pi*f_y*time_interval*time_array+phaseshift)
        z_t = x_t + y_t
    #put implementation with numpy over here 
    else:
        x_t = [A_x*math.cos(2*math.pi*f_x*i*time_interval) for i in range(num_iterations)]
        y_t = [A_y*math.sin(2*math.pi*f_y*time_interval*i+phaseshift) for i in range(num_iterations)]
        z_t = [x_t[i] + y_t[i] for i in range(num_iterations)]
    return x_t, y_t, z_t
```

```python
def saveAsASCII(x_t, y_t, z_t):
    x_t_converted = ascii(x_t)
    y_t_converted = ascii(y_t)
    z_t_converted = ascii(z_t)
    file_x = open("x_t_ASCII.txt", 'x')
    file_x.write(x_t_converted)
    file_y = open("y_t_ASCII.txt", 'x')
    file_y.write(y_t_converted)
    file_z = open ("z_t_ASCII.txt", 'x')
    file_z.write(z_t_converted)
```

```python
def plotFunctions (f_x, f_y, A_x, A_y, phaseshift, time_interval, num_iterations, useNumpy, filenum):
    x_t, y_t, z_t = compute_functions(f_x, f_y, A_x, A_y, phaseshift, time_interval, num_iterations, useNumpy)
    plt.plot(y_t, x_t) 
    plt.ylabel("x_t")
    plt.xlabel("y_t")
    plt.title("f_x / f_y = " + str(f_x/f_y))# + "  " + r'$\phi$ = ' + str(phaseshift))
    plt.savefig("fx%d"%(filenum))
    plt.show()
```
